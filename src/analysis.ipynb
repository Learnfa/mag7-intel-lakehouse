```python
import numpy as np
def analyze_price_brackets(df, ticker, target_date, lookback_window=200, analysis_window=100, num_brackets=10):
    """
    Splits recent price action into brackets and calculates average gain per bracket.
    """
    results = []
    
    target_date = pd.to_datetime(target_date)
    df = df.copy()

    df = df.reset_index()

    # Filter: Keep only data up to and including the target_date
    df = df[df['ticker'] == ticker]
    df = df[df['date'] <= target_date]

    # Ensure data is sorted
    df = df.sort_values(['ticker', 'date'])
    
    # Calculate day_gain if it doesn't exist yet
    if 'day_gain' not in df.columns:
        df['day_gain'] = df.groupby('ticker')['close'].diff()
    
    # Process each ticker separately
    for ticker, group in df.groupby('ticker'):
        # --- Step 1: Define the Range (Last 200 Days) ---
        subset_200 = group.tail(lookback_window)
        
        if len(subset_200) < 2:
            continue
            
        min_close = subset_200['close'].min()
        max_close = subset_200['close'].max()
        
        # Create 10 bins between min and max
        bins = np.linspace(min_close, max_close, num_brackets + 1)
        
        # --- Step 2: Analyze Performance (Last 100 Days) ---
        subset_100 = group.tail(analysis_window).copy()
        
        # Assign each row's Close price to a bracket
        # pd.cut creates categorical intervals
        subset_100['bracket'] = pd.cut(subset_100['close'], bins=bins, include_lowest=True)
        
        # --- Step 3: Aggregate ---
        # Group by bracket and calculate mean gain
        bracket_stats = subset_100.groupby('bracket')['day_gain'].agg(['mean', 'count'])
        
        # Cleanup for display
        bracket_stats['ticker'] = ticker
        results.append(bracket_stats)

    if not results:
        return pd.DataFrame()
        
    return pd.concat(results)
```

```python
# Run the analysis
analysis_result = analyze_price_brackets(df, ticker= 'AAPL', target_date='2024-01-01')

# Display result
print("Average Day Gain per Price Bracket (Past 100 Days Analysis):")
#tsla_stats = stats_on_date[stats_on_date['ticker'] == 'TSLA']
print(analysis_result)
```

```python
import pandas as pd
import numpy as np

def analyze_next_day_performance(df, ticker, target_date, lookback_window=200, analysis_window=100, num_brackets=10):
    """
    Calculates next-day predictive stats based on data available AS OF the target_date.
    
    Args:
        df (pd.DataFrame): Stock data.
        target_date (str): The 'current' date for the analysis (e.g., '2025-01-01').
        lookback_window (int): Days to look back for defining Min/Max range.
        analysis_window (int): Days to look back for calculating stats.
        num_brackets (int): Number of price bins.
    """
    results = []
    
    # 1. Prepare Date Filtering
    target_date = pd.to_datetime(target_date)
    df = df.copy()
    #df['date'] = pd.to_datetime(df['date'])

    # FIX: Reset index to ensure 'date' and 'ticker' are available as columns
    # This handles cases where you previously ran set_index(['ticker', 'date'])
    df = df.reset_index()
    
    # Filter: Keep only data up to and including the target_date
    df = df[df['ticker'] == ticker]
    df = df[df['date'] <= target_date]

    
    # Sort (critical for tail() and shift())
    df = df.sort_values(['ticker', 'date'])
    
    for ticker, group in df.groupby('ticker'):
        # --- Step 1: Define Brackets (Last 200 Days from target_date) ---
        subset_200 = group.tail(lookback_window)
        
        if len(subset_200) < 2:
            print(f"Not enough data for {ticker} before {target_date}")
            continue
            
        min_close = subset_200['close'].min()
        max_close = subset_200['close'].max()
        
        # Create bins
        bins = np.linspace(min_close, max_close, num_brackets + 1)
        
        # --- Step 2: Analyze Stats (Last 100 Days from target_date) ---
        subset_100 = group.tail(analysis_window).copy()
        
        # Calculate Next Day Gain
        # Note: The very last row (target_date) will be NaN because we don't know tomorrow's price yet.
        # This is correct for backtesting (no looking ahead).
        subset_100['next_day_gain'] = subset_100['close'].shift(-1) - subset_100['close']
        
        # Assign brackets
        subset_100['bracket'] = pd.cut(subset_100['close'], bins=bins, include_lowest=True)
        
        # --- Step 3: Aggregate ---
        bracket_stats = subset_100.groupby('bracket', observed=False)['next_day_gain'].agg(['mean', 'count'])
        bracket_stats['ticker'] = ticker
        results.append(bracket_stats)

    if not results:
        return pd.DataFrame()
        
    return pd.concat(results)
```

```python
# Analyze as if today is January 1st, 2021
stats_on_date = analyze_next_day_performance(df, ticker= 'META', target_date='2025-01-01')

#aapl_stats = stats_on_date[stats_on_date['ticker'] == 'AAPL']

# View the result
print(stats_on_date)

```

```python
import numpy as np
def analyze_combined_performance(df, ticker, target_date, lookback_window=200, analysis_window=100, num_brackets=10):
    """
    Analyzes next-day performance grouping by BOTH Price Range AND Fear/Greed Sentiment.
    """
    results = []
    
    # 1. Prepare Data
    df = df.copy()
    df = df.reset_index() # ensure columns are available
    #df['date'] = pd.to_datetime(df['date'])
    target_date = pd.to_datetime(target_date)
    
    # Filter & Sort
    df = df[df['ticker'] == ticker]
    df = df[df['date'] <= target_date]
    df = df.sort_values(['ticker', 'date'])
    
    # 2. Define Fixed Fear/Greed Bins
    # 0-33: Fear, 33-67: Neutral, 67-100: Greed
    fg_bins = [0, 33.33, 66.67, 100]
    fg_labels = ['Fear', 'Neutral', 'Greed']
    
    for ticker, group in df.groupby('ticker'):
        # --- Step A: Define Price Brackets (Last 200 Days) ---
        subset_200 = group.tail(lookback_window)
        if len(subset_200) < 2: continue
            
        min_close = subset_200['close'].min()
        max_close = subset_200['close'].max()
        p_bins = np.linspace(min_close, max_close, num_brackets + 1)
        
        # --- Step B: Prepare Analysis Set (Last 100 Days) ---
        subset_100 = group.tail(analysis_window).copy()
        
        # Calculate Next Day Gain
        subset_100['next_day_gain'] = subset_100['close'].shift(-1) - subset_100['close']
        
        # Assign Price Bracket
        subset_100['price_bracket'] = pd.cut(subset_100['close'], bins=p_bins, include_lowest=True)
        
        # Assign Fear/Greed Bracket
        subset_100['fg_bracket'] = pd.cut(subset_100['Fear_Greed'], bins=fg_bins, labels=fg_labels, include_lowest=True)
        
        # --- Step C: Aggregate by BOTH keys ---
        # Group by Price Bracket AND Fear/Greed
        stats = subset_100.groupby(['price_bracket', 'fg_bracket'], observed=False)['next_day_gain'].agg(['mean', 'count'])
        
        stats['ticker'] = ticker
        results.append(stats)

    if not results:
        return pd.DataFrame()

    return pd.concat(results)
```

```python
# 1. Run the analysis
combined_stats = analyze_combined_performance(df, ticker = 'AAPL', target_date='2025-01-01')

# 2. Pivot to create a readable matrix
# Rows = Price Brackets, Columns = Fear/Greed Levels
matrix = combined_stats.reset_index().pivot(
    index='price_bracket', 
    columns='fg_bracket', 
    values='mean'
)

# 3. Display with a heatmap style (optional)
# Using background_gradient makes the "Buy" (High +) and "Sell" (Low -) zones pop out
matrix.style.background_gradient(cmap='RdYlGn', axis=None).format("{:.2f}")
```
